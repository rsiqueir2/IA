```python
#Importações biblis;
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from pandas.io import gbq
```

```python
#Dataset com dados de venda de logos e seus respectivos lançamentos desde 1980 até 2020 rankeados por venda.
```

```python
#Carregar o dataset;
games=pd.read_csv('games.csv',encoding='UTF-8',sep=',')
```

```python
#Exibir dataset sem filtros;
#Colunas: Rank, Nome, Plataforma, Ano de lançamento, Gênero, Publicado por, America do Norte vendas, Europa Vendas, Japão Vendas
# Outras vendas além dos 3 continentes citados e Vendas Global.
games
```

```python
#Todos os modelos de Graficos Pairplot
sns.pairplot(games)
```

```python
#Exibindo o Rank baseado na qtde de venda e o ano que foi lançado.
g = sns.lmplot(x="Year", y="Global_Sales", hue="Year",
               truncate=True, height=5, data=games)
```

```python
#Agrupando por ano de lançamento e nome dos jogos;
grupAno = games.groupby(['Year','Name']).sum()
```

```python
#Exibindo os dados do filtro;
grupAno
```

```python
#Filtrando por ano 1985 e Rank 2 = Mario ♥;
games[(games['Year'] == 1985) & (games['Rank'] == 2)]
```

```python
#Filtrando por ano e Genero;
games[(games['Year'] == 2000) & (games['Genre'] == 'Platform')]
```

```python
#Exibindo 
g = sns.lmplot(x="Year", y="Global_Sales", hue="Platform",
               truncate=True, height=7, data=games)
```
